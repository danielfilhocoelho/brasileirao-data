```python
from bs4 import BeautifulSoup
import requests
import pandas

headers = {'User-Agent': 
           'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/47.0.2526.106 Safari/537.36'}

years = ["2005","2006","2007","2008","2009","2010","2011","2012","2013","2014","2015","2016","2017"]


for year in years:
    page = 'https://www.transfermarkt.com.br/campeonato-brasileiro-serie-a/formtabelle/wettbewerb/BRA1?saison_id=' + year + '&min=1&max=38'
    site = requests.get(page,headers=headers)
    bs = BeautifulSoup(site.content,'html.parser')
    table = bs.find('div', {'class' : 'responsive-table'})
    x = pandas.read_html(table.prettify())
    tabela_brasileirao = x[0]
    del(tabela_brasileirao['Clube'])
    del(tabela_brasileirao['#'])
    tabela_brasileirao = tabela_brasileirao.rename(columns={"Unnamed: 2": "Clube", "Unnamed: 9": "Pontos", "V":"Jogos" , "E":"V", "D":"E","Golos":"D","+/-":"Gols pr√≥/contra","Pts":"Saldo"  })
    true_year = int(year) + 1
    tabela_brasileirao.to_csv(r'tabela' + str(true_year) + '.csv',index=False)


```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


years = ["2006","2007","2008","2009","2010","2011","2012","2013","2014","2015","2016","2017","2018"]
tabelas = []
for year in years:   
    tabelas.append(pd.read_csv('tabela' + year + '.csv' ))
    
vitorias = []
count = 0

while (count <= 19):
    vitorias = []
    for t in tabelas:
        vitorias.append(t['V'][count])
        plt.hist(vitorias,align='left',bins=np.arange(np.min(vitorias),np.max(vitorias)+2),rwidth=0.5,color = 'blue')
        plt.title("Hist√≥rico de vit√≥rias do "+ str(count + 1) +" colocado de 2006 a 2018" )
    plt.show()
    count = count + 1
```
